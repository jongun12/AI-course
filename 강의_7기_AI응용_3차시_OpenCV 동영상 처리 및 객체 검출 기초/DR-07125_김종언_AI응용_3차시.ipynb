문항 3. 동영상 객체 생성 및 예외 처리 bird.mp4 파일을 읽어오는 VideoCapture 객체 cap을 생성하고, 파일이 정상적으로 열리지 않았을 경우 에러 메시지를 출력하고 종료하는 코드를 작성하세요.

```python
import cv2

video_path = r'/home/kim/Desktop/AI_cours/data/bird.mp4'

# TODO: VideoCapture 객체 생성
cap = cv2.VideoCapture(video_path)

# TODO: 파일이 열렸는지 확인 (열리지 않았으면 에러 출력)
if not cap.isOpened() :
    print("ERR : 파일을 열 수 없습니다.")
    exit()
```

문항 4. 동영상 속성 정보 확인 위에서 생성한 cap 객체를 이용하여 동영상의 초당 프레임 수(FPS), 가로 폭(Width), 세로 높이(Height) 정보를 가져와 출력하는 코드를 작성하세요. (힌트: cap.get() 메서드와 cv2.CAP_PROP_ 상수를 사용하세요.)

```python
# TODO: FPS, Width, Height 속성 가져오기
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

print(f'fps: {fps}, width: {width}, height: {height}, count: {frame_count}')
```

문항 5. 프레임 읽기 및 그레이스케일 변환 while 반복문을 사용하여 동영상의 프레임을 한 장씩 읽어오고(read), 읽어온 프레임이 있다면 이를 **그레이스케일(Grayscale)** 로 변환하여 cv2.imshow로 출력하는 코드를 작성하세요. (종료 조건은 다음 문제에서 작성합니다.)

```python
while True:
    # TODO: 프레임 읽기 (ret: 성공여부, frame: 이미지데이터)
    result, frame = cap.read()

    # 프레임을 더 이상 읽을 수 없으면(동영상 끝) 반복 종료
    if not result:
        break

    # TODO: 읽은 프레임을 흑백(Grayscale)으로 변환
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    # 화면 출력
    cv2.imshow('bird',gray)  # 콜랩에서 이미지 출력

    # q를 누르면 종료 (콜랩의 getch 대체)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```
